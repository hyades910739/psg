## 繼承路徑
* 由下至上，由左至右
* I2.w 的搜索路徑為 I2, C1, C2, C3 (I2 從 C3 繼承了屬性 w)

## 類方法調用
* 如果I2.w引用是一個函數調用，其實際的含義是 “條用C3.w函數處理I2“
* I2.w() == C3.w(I2)

## 編寫類樹

```
class C2: ...

class C3: ...

class C1(C2, C3): ...

I1 = C1()
I2 = C1()
```

屬性通常是在類內對傳給函數的特殊參數(self)做賦值運算添加在實例中

```python
class C1(C2, C3):
    def setname(self, who):
        self.name = who
```
```python
I1 = C1()
I2 = C1()
I1.setname('Louis')
I2.setname('Rita')
print(I1.name)  # get Louis
```

目前寫的 code ，直到調用 setname 方法前，C1 類都不會把 name 屬性附加在實例之上，除非...
```python
class C1(C2, C3):
    def __init__(self, who):
        self.name = who
        
I1 = C1('Louis')
I2 = C1('Rita')
print(I1.name)  # get Louis
```

## OOP 是為了代碼重用
#### 例子：假如今天幫公司寫一個可以有效管理員工的程式，可能會...

```
class Employee:
    def computeSalary(self): ...
    def giveRaise(self): ...
    def promote(self): ...
    def retire(self): ...
```
（Employee 為一個通用的超類，假如沒有特別聲明的員工 則利用以上方法調用）

當有特殊員工需要有特別的計薪方式時...
```python
class Engineer(Employee):
    def computeSalary(self):
```

### example
Louis 為一個很普的員工
```python
louis = Employee()
```
Rita 為一個工程師
```python
rita = Engineer()
```

### 總結：OOP 就是代碼的重用，最小化程式的冗余
